#### Load data train and result prediction


```python
df_train = pd.read_csv("data/all_category/data_run/_1M_shopee/train_set/shopee_train.csv", sep='\t')
# df_train = pd.read_csv("data/all_category/data_run/_30k_kiot/all_cate_data_train.csv", sep='\t')
# df_infer = pd.read_csv("data/all_category/data_run/_1M_shopee/test_set/kiotviet_test_predicted.csv", sep='\t')
df_infer = pd.read_csv("data/all_category/data_run/_1M_shopee/test_set/official_test/_2k5_labeled_predicted.csv", sep='\t')
```

```python
df_infer
```

#### Confusion report

```python
from sklearn.metrics import confusion_matrix
y_true = df_infer["label"].values
y_pred = df_infer["predicted_label"].values
labels = list(set(y_pred))
cnf_matrix = confusion_matrix(y_true, y_pred, labels=labels)
print(labels)
```

```python
cate="Điện tử - Điện máy"

confusion_report(
    cate=cate,
    cnf_matrix=cnf_matrix,
    labels=labels,
    df_train=df_train)
```

#### False positive

```python
fp_set = df_infer.loc[np.logical_and(df_infer["label"]!=cate,df_infer["predicted_label"] == cate)]
print(f"Number false positive: {len(fp_set)}") 
fp_set
```

#### False Negative


```python
fn_set = df_infer.loc[np.logical_and(df_infer["label"]==cate,df_infer["predicted_label"] != cate)]
print(f"Number false negative: {len(fn_set)}")
temp = fn_set[fn_set.predicted_label == "Thời trang"]
temp.iloc[:50]
```
